Problem 15:  Create three variables: product = 'Laptop', price = 75000.5, discount = 10. Print a formatted line using all three string formatting styles (f-string, .format(), % operator) showing: 'Product: Laptop | Price: 75000.50 | Discount: 10%'.

Approach:
I used three different string formatting methods : f-string, '.format()', and '%' formatting to display product information in a well-formatted and readable way. The price was formatted to show two decimal places.


```python
product = "Laptop"
price = 75000.5
discount = 10

print(f"Product: {product} | Price: {price:.2f} | Discount: {discount}%")

print(
    "Product: {} | Price: {:.2f} | Discount: {}%".format(
        product, price, discount
    )
)

print(
    "Product: %s | Price: %.2f | Discount: %d%%"
    % (product, price, discount)
)
```

Problem 18:  Take a score (0 - 100) from the user and print a grade using if-elif-else: 90–100 → 'A+', 80–89 → 'A', 60–79 → 'B', 40–59 → 'C', below 40 → 'F'. Also print 'Pass' or 'Fail' based on score >= 40.

Approach:
I used an if-elif-else statement to determine the grade based on the student's score. Then, a separate condition was used to check whether the student passed or failed.

```python
score = int(input("Enter score: "))

if 90 <= score <= 100:
    grade = "A+"
elif 80 <= score <= 89:
    grade = "A"
elif 60 <= score <= 79:
    grade = "B"
elif 40 <= score <= 59:
    grade = "C"
else:
    grade = "F"

print("Grade:", grade)

if score >= 40:
    print("Pass")
else:
    print("Fail")
```

```python
correct_username = "admin"
correct_password = "12345"

username = input("Enter username: ")
password = input("Enter password: ")

if username == correct_username:

    if password == correct_password:
        print("Access Granted")
    else:
        print("Wrong Password")

else:
    print("User Not Found")
```

Problem 20:  Take any number from the user. Using a single line ternary expression, print whether it is 'Positive', 'Negative', or 'Zero'. Then separately, using nested if-elif-else, also check if the number is even or odd (only if it is not zero).

Approach:
I used a nested ternary expression to determine whether the entered number is positive, negative, or zero. Then, used conditional statements to check whether the number is even or odd if it is not zero.

```python
number = int(input("Enter a number: "))

result = (
    "Positive"
    if number > 0
    else ("Negative" if number < 0 else "Zero")
)

print(result)

if number != 0:

    if number % 2 == 0:
        print("Even")
    else:
        print("Odd")

else:
    print("Zero is neither even nor odd.")
```
